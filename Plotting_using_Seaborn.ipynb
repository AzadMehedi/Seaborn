### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# scatterplot   -> axes label function
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
# line plot
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue  ->  style
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp',kind='line', hue='country')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line', hue='country', style='continent')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter', row='smoker', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter', row='sex', col='day')
```

```python
sns.relplot(data=tips, x='total_bill',y='tip',kind='line', row='smoker', col='sex')
```

```python
# col wrap
gap
```

```python
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
tips
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
# figure level -> displot
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
# Its also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables
# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# element -> step
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic, x='age', kind='hist', hue='sex', col='survived', element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=titanic, x='age', kind='hist', col='sex', element='step')
```

```python
# kdeplot - kernel density estimation plot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
tips.head()
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```
